```python
import torch
import os

# Load the projection matrix and concept files
t_W_c = torch.load(os.path.join(t_dir, 'W_c.pt'), map_location='cpu')  # Loading the weight matrix
# Ns = 576  # Number of spatial concepts
# Nt = 305  # Number of temporal concepts

# Loading spatial and temporal concept names
with open(os.path.join(s_dir, 'concepts.txt')) as f:
    s_concepts = f.read().splitlines()

with open(os.path.join(t_dir, 'concepts.txt')) as f:
    t_concepts = f.read().splitlines()

# Make sure t_W_c, s_concepts, and t_concepts match their expected sizes
assert t_W_c.shape == (Nt, Ns), "Mismatch in dimensions between t_W_c and concept sizes"

# Find top 3 spatial concepts for each temporal concept
top_spatial_concepts = {}

for t_idx in range(Nt):
    # Get the top 3 indices and values for the current temporal concept
    top3_values, top3_indices = torch.topk(t_W_c[t_idx], 10)
    
    # Get the corresponding spatial concept names, rounding values to 2 decimal places
    top3_concepts = [(s_concepts[i], round(top3_values[j].item(), 2)) for j, i in enumerate(top3_indices)]
    
    # Store the results
    top_spatial_concepts[f"Temporal {t_idx + 1}:{t_concepts[t_idx]}"] = top3_concepts

# Print the results
for t_concept, top3 in top_spatial_concepts.items():
    print(f"{t_concept}:")
    for i, (s_concept, value) in enumerate(top3):
        print(f"  - Spatial {i + 1}: {s_concept} (Value: {value:.2f})")
```
